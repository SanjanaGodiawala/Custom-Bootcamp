```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x_axis = np.array([1,8,10,])
y_axis = np.array([1,25,50])
```

```python
plt.plot(x_axis,y_axis)
```

```python
y_points=np.array([1,10,2,8,5])
plt.plot(y_points)
```

```python
y_points=np.array([1,10,2,8,5])
plt.plot(y_points,marker="o")
```

```python
y_points=np.array([1,10,2,8,5])
plt.plot(y_points,marker="*", color ="purple" , linestyle="dotted")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Sample Graph")
```

```python
Create a list of months: ['January', 'February', 'March', 'April', 'May'].
Create a list of monthly sales data for a hypothetical product: [2500, 3200, 2800, 4100, 3700]
    
```

```python
Months=np.array(['January', 'February', 'March', 'April', 'May'])
Sales = np.array([2500, 3200, 2800, 4100, 3700])
```

```python
plt1=plt.bar(Months,Sales)
```

```python
plt.bar(Months,Sales,color="Purple")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly sales data")
plt.savefig("Dataset/Monthly_Sales.png")
```
